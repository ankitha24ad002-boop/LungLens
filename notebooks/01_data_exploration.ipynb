# LungLens - Data Exploration

## Objective
Explore and understand the lung image dataset before preprocessing.

## Tasks:
- Check dataset structure
- Count image classes
- Check image dimensions
- Visualize sample images

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
data_path = "../data/raw"

images = os.listdir(data_path)

print("Total images:", len(images))
print(images[:10])
```

```python
data_path = "../data/raw"

images = os.listdir(data_path)

print("Total images:", len(images))
print(images[:10])
```

```python
bacteria = []
virus = []

for img in images:
    if "BACTERIA" in img:
        bacteria.append(img)

    elif "VIRUS" in img:
        virus.append(img)

print("Bacteria images:", len(bacteria))
print("Virus images:", len(virus))
```

```python
plt.figure(figsize=(12,6))

for i, img_name in enumerate(images[:6]):

    if img_name.endswith(".jpeg"):

        img = cv2.imread(
            os.path.join(data_path, img_name)
        )

        img = cv2.cvtColor(
            img,
            cv2.COLOR_BGR2RGB
        )

        plt.subplot(2,3,i+1)
        plt.imshow(img)
        plt.title(img_name)
        plt.axis("off")

plt.show()
```

```python
for img_name in images[:5]:

    if img_name.endswith(".jpeg"):

        img = cv2.imread(
            os.path.join(data_path, img_name)
        )

        print(img_name, "->", img.shape)
```

```python
sample_images = bacteria[:3] + virus[:3]

plt.figure(figsize=(12,6))

for i, img_name in enumerate(sample_images):

    img = cv2.imread(
        os.path.join(data_path, img_name)
    )

    img = cv2.cvtColor(
        img,
        cv2.COLOR_BGR2RGB
    )

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title(img_name)
    plt.axis("off")

plt.show()
```
