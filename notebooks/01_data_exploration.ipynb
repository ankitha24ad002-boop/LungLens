```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

dataset_path = "/content/drive/MyDrive/LungLens/data/processed/train"

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

classes = ["BACTERIA", "VIRUS"]

for row, cls in enumerate(classes):

    class_path = os.path.join(dataset_path, cls)

    images = [
        f for f in os.listdir(class_path)
        if f.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    selected = random.sample(images, min(2, len(images)))

    for col, image_name in enumerate(selected):

        image_path = os.path.join(class_path, image_name)

        image = Image.open(image_path)

        axes[row, col].imshow(image)
        axes[row, col].set_title(image_name)
        axes[row, col].axis("off")

plt.tight_layout()
plt.show()
```

```python
import os

total = 0

for split in ["train", "val", "test"]:

    split_path = os.path.join(dataset_path, split)

    print(f"\n{split.upper()}")

    for cls in os.listdir(split_path):

        class_path = os.path.join(split_path, cls)

        count = len([
            f for f in os.listdir(class_path)
            if f.lower().endswith((".jpg", ".jpeg", ".png"))
        ])

        print(f"{cls}: {count}")

        total += count

print("\nTotal Images:", total)
```
